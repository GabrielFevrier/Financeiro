```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas_ta as ta
```

```python
data = pd.read_csv('WIN_M10.CSV')
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m-%d %H:%M')
data.index = data.pop('Date')
data
```

```python
data['RSI']=ta.rsi(data.Close, length=15)
data['EMAF']=ta.ema(data.Close, length=20)
data['EMAM']=ta.ema(data.Close, length=100)
data['EMAS']=ta.ema(data.Close, length=150)
data['Target'] = data['Close']-data.Open
data['Target'] = data['Target'].shift(-1)
data.dropna(inplace=True)
data
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data[['Bar', 'Open', 'High', 'Low', 'Close', 'Volume', 'RSI', 'EMAF', 'EMAM', 'EMAS', 'Target' ]])
```

```python
sequence_length = 30
```

```python
def create_sequences(data, seq_length):
    sequences = []
    for i in range(len(data) - seq_length):
        sequences.append(data[i:i+seq_length])
    return np.array(sequences)
```

```python
X = create_sequences(scaled_data, sequence_length)
```

```python
y = (data['Close'].shift(-1) > data['Close']).iloc[:-sequence_length].astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=28)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.LSTM(150, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    tf.keras.layers.LSTM(75, return_sequences=True),  # Adicionando outra camada LSTM
    tf.keras.layers.LSTM(75,  return_sequences=True),
    tf.keras.layers.LSTM(75),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=32)
```

```python
predictions = model.predict(X)
```

```python
loss, accuracy = model.evaluate(X, y)
print(f'Test Loss: {loss}, Test Accuracy: {accuracy}')
```
